```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_path = "G:/DATA ANALYTICS/Projects/kindel dataset/kindle_data.csv"
df = pd.read_csv(dataset_path)
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
print(df.isnull().sum())
df=df.dropna()
```

```python
print(df.describe())
```

```python
# Histogram for a numerical column (e.g., 'stars')
```

```python
plt.hist(df['stars'],bins=10,edgecolor='black')
plt.xlabel('Stars')
plt.ylabel('Frequency')
plt.title('Distribution of Stars')
plt.show()
```

```python
# Countplot for a categorical column (e.g., 'category_name')
```

```python
sns.countplot(x='category_name',data=df)
plt.xticks(rotation=90)
plt.xlabel('Category_name')
plt.ylabel('Count')
plt.title('Distribution of Categories')
plt.show()
```

```python
# Scatter plot for two numerical columns (e.g., 'stars' vs 'reviews')
```

```python
sns.scatterplot(x='stars',y='reviews',data=df)
plt.xlabel('Stars')
plt.ylabel('Reviews')
plt.title('Scatter Plot: Stars vs Reviews')
plt.show()
```

```python
# Boxplot for a numerical column and a categorical column (e.g., 'stars' vs 'category_name')
```

```python
sns.boxplot(x='category_name',y='stars',data=df)
plt.xticks(rotation=90)
plt.xlabel('Category Name')
plt.ylabel('Stars')
plt.title('Boxplot: Stars by Category')
plt.show()
```

```python
# Select only numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix for numeric columns
correlation_matrix = numeric_columns.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
